```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
file_sizes = {}
```

```python
def get_file_sizes(d):
    size_arr = []
    
    with os.scandir(d) as files:
        for f in files:
            size_arr.append(os.path.getsize(f))
    
    return np.array(size_arr)
```

```python
for d in os.listdir():
    if os.path.isdir(d):
        file_sizes[d] = get_file_sizes(d)
print(file_sizes)
```

```python
for d in file_sizes.keys():
    plt.figure()
    plt.boxplot(file_sizes[d])
    plt.title(d)
```
