```python
import pathlib
import os
import glob
```

Number each image in each category (e.g. leopard_000, leopard_001, ..., leopard_999)

Code inspiration: https://stackoverflow.com/questions/17748228/rename-multiple-files-in-python

```python
# Testing (Works!)
i = 0
for file in glob.iglob('dataset/snow leopard copy/*'):
    os.rename(file, 'dataset/snow leopard copy/snow leopard_{:0>3d}.jpg'.format(i))
    i += 1
```

```python
glob.glob('dataset/snow leopard copy/*')
```

```python
categories = ['cheetah', 'jaguar', 'leopard', 'snow leopard']

for cat in categories:
    i = 0
    for file in glob.iglob('dataset/{}/*'.format(cat)):
        os.rename(file, 'dataset/{}/{}_{:0>3d}.jpg'.format(cat, cat, i))
        i += 1
```
