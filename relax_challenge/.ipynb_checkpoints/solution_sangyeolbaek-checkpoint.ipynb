```python
user_ids  = user_adopter.user_id.unique()
user_logs = {i: np.array(user_adopter_new.loc[i].index.tolist()) for i in user_ids}
user_logs
```

```python
def is_adopter(times):
    target_delta = np.timedelta64(3, 'D')
    
    for i in range(len(times)-2):
        if np.datetime64(times[i+2], 'D') - np.datetime64(times[i], 'D') <= target_delta:
            return True
    return False
```

```python
print(is_adopter(user_logs[10]))
```

```python
user_adopter = {i: is_adopter(user_logs[i]) if len(user_logs[i]) >= 3 else False for i in user_logs.keys()}
user_adopter
```

## Conclusion

This challenge was to find which factors would be the most important in determining whether a user would be an "adopted user." For this, I created some new columns derived from the original datasets, namely:
- `creation_time_float`, which tells how long one has been a user in seconds
- `total logs`, which tells number of occurrences for each `user_id` in **takehome_user_engagement.csv**
- `creation_time_float` and `creation_source_num` were created due to requiring numeric values to the model

Some factors I considered but omitted due to time constraint were the users' maximum numbers of days spent logged in a 7-day period, and the user's average login rate in respect to their duration as a user.

From the graph, it seems that the user's `total_logs` is the most important feature by a mile, with a score of 0.91. The next significant factors contribute less than 0.02, with some of them related to their organization's id. However, `total_logs` still dwarf the other features in importance, making it the best predictor for a user's "adopted user" status.
